```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import math
```

```python
fig = plt.figure(figsize=(30, 10))
ax = plt.axes()
alphas = [0.1, 0.2]
betas = [0.1, 0.2]
c = 1000
for alpha, beta in zip(alphas, betas):
#     ctr = np.linspace(0, 1, 100) * 0.01
    ctr = [0.8 for _ in range(c)]
    cvr = np.linspace(0, 1, c)

    score1 = np.power(ctr, alpha) + np.power(cvr, beta)
    score2 = np.power(ctr, alpha) * np.power(cvr, beta)
    plt.plot(range(c), score1, '-', label="alpha={}_beta={}".format(alpha, beta));
    plt.plot(range(c), score2, 'o', label="alpha={}_beta={}".format(alpha, beta));
plt.legend()
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
p = np.linspace(0.001, 0.2, 1000) 
score = p*1.0 / (1-p)
plt.plot(p, score, '-', label='score') 
plt.plot(p, p, '--', label='p') 
plt.legend();
```
